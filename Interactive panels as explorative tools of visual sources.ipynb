In this article we propose a methodology and a tool for exploring visual collections, based on the interrelations of historical and iconographic connections, through the use of interactive storytelling. The methodology in question is strongly inspired by the work of the art historian Aby Warburg and, most especifically, by his Atlas Mnemosyne project; in this line we propose a way of using an open source, bilingual, JavaScript library called Aventura.js, developed by one of the authors, to create interactive iconographic panels organized in semantic spaces.
Following the structure of the Journal, the article is organized in three overlapping and intertwining layers: the narrative layer of the article presents the context and the epistemological principles originally proposed by Warburg on his Atlas Mnemosyne project at the beginning of the XX century, and also to reflect on new ways to study visual source materials by using digital tools that help to algorithmically structure the semantic relationships between images and to interactively allow an audience to explore them following the insights of warburg. We will discuss the need to develop specific tools to visualize qualitative data in image corpuses in humanities and social sciences. Concretely, panels organized in 2D space according to categories related to the persistence of iconographic motifs in collective memory, but, additionally, organized in 3D space, which we could call the hipertextual z-axis, in terms of depth of meaning elaboration and intertextuality. ## Atlas Mnemosyne after digital humanities

The authors of this article fortuitously came across a happy coincidence in terms of the topic presented. We both saw in the epistemological principles of Aby Warburg's Mnemosyne Atlas (Warburg, 2012), the potential to address the concern for the creation of digital tools to visualize visual sources, motivated by the limitations of conventional publication formats (Urueña Calderón, 2017) and by the need to adequately visualize the categorical data used in the humanities (Gómez, 2021)<cite data-cite="7299034/779MQTKK"></cite>.
In Warburg's words, the Atlas Mnemosyne was an instrument of "orientation" that from the exposition of "iconic materials" and their visual relationships allowed to "illustrate" the "process" of how in the Renaissance the values of expression of "life in motion" coined in antiquity were incorporated (Warburg, 2012, p. 35). The final form that the Atlas took consisted of the use of large black fabric screens attached to frames measuring 1.50 m by 2 m, on which different photographs could be grouped by attaching them by means of small, easily manipulated clamps (Didi-Huberman, 2009, p. 412). The panels that make up the Atlas consist of a set of images organized by means of visual relationships defined by certain recurring motifs of themes, gestures and bodily dispositions found in a heterogeneity of sources not considered, at the time, by art historians. All this material was organized in groups and series, according to the links suggested by the different visual, historical and geographical relationships. This procedure has been described as a montage work similar to what Sergey Eisenstein or Dziga Vertov were developing at the time (Didi-Huberman, 2009; Michaud, 2004; Urueña, 2017). Nowadays, it has also served to think about the ways of collecting, safeguarding, interpreting, arranging and exhibiting visual materials in the digital world (Preez, 2020).
The procedure to organize and manipulate the elements of the collection that integrates the Atlas is related to the epistemological principle of the "good neighbor". This implies that it is possible to creatively explore the archives and libraries by finding surprising relationships because of the significant neighborliness of elements that are next to those we are looking for. The principle of good neighborliness was caused by Warburg because of the special spatial arrangement of his grand library (Settis et al., 2010), and served as an inspiration for imagining modes of organization and arrangement of, and relationships between, digital resources of online catalogs and databases of museums, archives, libraries, and other heritage organizations (Brandhorst, 2013; Patti & Quiviger, 2014; Warnke & Dieckmann, 2016).
Good neighborliness is not founded on chance, but on a special consideration of the relationships between the particular and the general. Between the elements of the collection, relationships are not established by induction or deduction, but by an exemplary or paradigmatic logic, in which they feed each other according to the meaningful neighborliness between singular manifestations of the same theme or iconographic detail. The units of meaning of the collection are called Pathosformel, or expressive formula. This concept refers to the expression by visual means of intense emotions, or Pathos, which are manifested plastically in gestures or bodily dispositions. These motifs become formulas that, when copied and used as models, are resignified and adapted to new historical and geographical contexts. The repetitive nature of the Pathiosformel has even allowed the construction of algorithms that recognize certain poses or gestures (Impett, 2020).
The circulation of these visual materials in time and space has a behavior that Warburg calls survival (Nachleben), which designates a state of latency in which the motifs represented in the images, or Pathosfomel, are maintained as they move between periods, epistemic domains, contexts of signification, supports, techniques and places through copying procedures which result in dramatic inversions of meaning. Each time these motifs come into contact with an artist, an spectator or a scholar they are "recharged with life" and put to work in new contexts. This contact is characterized by Warburg as a "confrontation" in which artists, spectators and scholars struggle between surrendering thoughtlessly to the power of the image, or, on the contrary, establishing a reflexive distance (Denkraum).
In this way, Mnemosyne is an orientation tool that enables scholars, artists and spectators to establish reflexive distance (Denkraum) and to make comparisons between materials in a working "space" in which different meaningful juxtapositions can be tried out. The interval (Zwichenraum) is precisely one of the central concepts of the Atlas. In Mnemosyne such an interval is graphically achieved by the black background fabric, which establishes a space between different frames. In a context in which the use of photographs and film projectors for teaching and disseminating art history research had not yet become popular, Warburg formed a large collection of photographs of the different materials he was interested in exploring (Mazzucco, 2012, 2013). These were organized in series and were placed on the same epistemological level by being placed on the black fabric panels (Michaud, 2004, p. 278). These panels were then re-photographed in order to create a single image, which was inserted into a series that was intended to take the form of a book.
The Mnemosyne Atlas has raised obvious comparisons between the screen of different devices and the workspace of the black fabric panels. In this context the concept of framing has been used to show relationships between the analog and digital worlds. The digital screen and fabric panel frame a selection of images and enable the comparative consideration of the chosen images in proximity and make possible a digital "Interval" (Zwichenraum), in which the potential originally expressed in Warburg's project of navigating between hyperlinks is realized (Doulkaridou, 2015; Preez, 2020).
Likewise, the comparison can be useful with the Cartesian space of the “Canvas”. The arrangement of the elements according to "x" and "y" coordinates allows the construction of a characterized space (features spaces) (Manovich, 2015) in which the shapes of the data can be compared depending on the chosen categories. Additionally, the different layouts, or panels, can be organized as scenes that are navigated by means of the z-axis of depth in routes made possible by the concept of hypertextual narrative (Landow, 2009).
Thus we will provide an example which consists of a collection of images that are iconographically related to the figure of the "Carguero" (Man carrier), a typical character of the 19th century Colombian visual culture. This figure could be interpreted as a concrete and localized manifestation of the Pathosformel of carrying something or someone. Although it would be more accurate to say that it is an iconographic theme, which takes the form of the representation of a character or "social type,” whose variations are structured by various details, some of which can be considered Pathosformeln (gestures and bodily dispositions). The images of this character use previous models, which, through repetition, have become representational formulas expressing intense emotions about the use of men for practices that are usually carried out by pack animals. The emotions, evaluations, experiences, and expectations that condense these representations, which can be traced to different associated textual sources, are specific historical manifestations. They account for a series of transformations related to the emergence of the modern world, such as the circumstances in which the industrial revolution, scientific revolutions, and political revolutions of the Atlantic world manifested themselves in the Latin American region. The ways in which these images are interpreted vary according to how they are situated in specific historical and geographical contexts, as well as in communities of interpretation related to travel literature, costume books, geographical relations, the literature of customs, the local press, and the international illustrated press.
The collection of images was gathered through a search in heterogeneity of analog and digital repositories and archives, based on Warburg's principle of good neighborliness. In this sense, it confronted disciplinary barriers that separate the images from the texts to which they are related and distribute them in different collections guarded, annotated, and cataloged by professionals from different disciplines. For example, in the Museo Nacional de Colombia, watercolors and loose plates can be found in the art collection, while engravings included in travel books are in custody of the history collection. This is a problem that is repeated in most of the repositories visited, in which visual materials receive different treatments: when they are in books, the images are ignored, and when the images are exhibited, the books, newspapers, and magazines they come from are ignored. This creates a series of problems in dealing with search strategies through queries and keywords, which must also be faced with the problem of the heterogeneity of classification systems and the language used when searching through archives in France, the United States, England or Germany. In addition, the collection also considers a number of popular uses of these images that are not held in repositories but are recorded by means of photographs and added by the authors directly to the collection. Thinking about the use of digital archives and their dialogue with analog and other types of unconventional and popular everyday methods of collecting and collecting both benefits from and have the potential to benefit from reflections on the use of the good neighbor principle to explore digital collections (Brandhorst, 2013; Patti & Quiviger, 2014; Preez, 2020).
It is also necessary to consider the use of digital image formats to homogenize the materials included in the collection. Just as Warburg used photographs to epistemologically equalize images of different formats and techniques (Michaud, 2004, p. 278), researchers working with digital images keep in their folders, organized through different categories, images in digital formats such as PNG and JPG, which they have used to homogenize a collection of elements coming from heterogeneous sources in order to stabilize them and put them in levels of similarity that allow comparisons, juxtapositions and contrasts. As Johanna Drucker has shown, these operations involve unavoidable mediations between visual sources and their digital equivalents, which take place even in the cases of originally digital sources, since between copies, compressions and format changes there are always transformations that can affect computational analyses (Drucker, 2013). The "Carguero" (Man carrier) image collection is made up of images which were photographed or scanned and then converted to PNG format from physical works such as bookplates, paintings, murals, sculptures, graffiti, stickers, as well as from different digital repositories that host books, magazines, newspapers, loose sheets, paintings, engravings, etc. in JPG, PDF and other formats.
In the first stage, these images were put in relation to each other in intuitive ways in procedures that tried to emulate analog forms. One of the first solutions proposed to adapt Warburg's panels to digital environments was to create panels with "black fabric" like backgrounds in Photoshop, in which the images were arranged intuitively according to the spatial relationships suggested by the categories of analysis. In a second attempt, the Aventura.js library made it possible to adapt the panels made in Photoshop by placing "areas" on the images, which allowed them to be navigated through scenes in iconological narratives. The new version presented here is an attempt to streamline the process by creating a CSV list in which the images were annotated using subjective categorical metadata, indicated by the research question, which will be discussed later in this text. In this way, they can be loaded, manipulated, and visualized with programming tools provided by libraries such as Python's Pandas, Matplotlib, and JavaScript's d3. 
Thus, Warburg's montage was adapted to the space of the digital screen. More specifically, to the HTML element "Canvas,” in whose Cartesian space, according to the "x" and "y" values, the images are spatially distributed according to the relationships of the categories previously noted in the CSV list. These arrangements form a "feature space" (Manovich, 2015), or an "interval" (Zwichenraum) (Warburg, 2012), that makes it possible to establish a “space for reflection” (Denkraum) and to find similarities and differences through comparison. These graphs, which we will explore in more detail below, are exported to Aventura.js and organized into scenes and areas that allow the construction of interactive hypertextual narratives. This heuristic tool has made it possible to indicate routes of analysis, add nuances, reaffirm indications, reformulate hypotheses, and ultimately refine and visualize the categorical apparatus that organized the material and guided the development of the historical narrative.

## Creating interactive panels

The method we used for constructing our interactive panels can be summarized in the following steps and associated tools:

1)	Preparing the subjective collection dataset, using conventional table-based representations, in CSV format, and online hosted images.

2)	Creating data subsets by filtering the main dataset, using the Pandas library in Python and off the shelf functions in JavaScript.

3)	Displaying subsets in adequate graphic representations, using the d3 visualization software for JavaScript in the Observable coding notebooks platform.

4)	Adding a narrative Z-dimension that connects all the graphic representations, using the Aventura software for creating interactive stories.

Most of the tools used in this project are rather conventional and well known, however, the main innovation in our project in terms of tools is the use of the Aventura software for explorative storytelling. Aventura is a JavaScript library, intended for the creation of electronic literature and generative art. In other words, works that use computation to help the production of visual and textual elements or to structure the navigation of a narration. Aventura provides a set of functions with which it is possible to create generative texts and images (using Markov chains and Context Free Grammars), and interactive stories (through a system for designing hypertextual, interconnected scenes). From the start, Aventura was developed as a bilingual (Spanish and English) alternative to other similar libraries, with the purpose to address idiomatic gaps in computational creativity, and to allow Spanish speaking users to make these types of creative works in their own language. As Takhteyev (2012) <cite data-cite="7299034/RF6KHEXQ"></cite> has studied, the practice of software development in Latin America occurs in a state of “cultural diglossia”, existing between two languages at the same time, the default language of coding and software documentation: English, and other local languages: for instance, Spanish or Portuguese. In this sense, addressing the linguistic gap by creating local software helps to reappropriate computational tools for the study of humanities in situated contexts and to reduce knowledge gaps derived from linguistic barriers.

As was described, Aventura is a tool for creative coding, but it can be used as well for exploratory research in the humanities and for divulgating the results of such research. One of the underlying propositions of this paper is that strategies from computational creativity can be used as research and communication methods. Furthermore, in this paper we want to focus on the potential of using Aventura’s interactive story system for creating iconographic panels, inspired by Warburg’s Atlas Mnemosyne, and the possibility of extending Warburg’s system of interpretation by means of data visualization and storytelling. We, thus, used Aventura to design a narrative exploration of the iconographic collection of the Cargueros image by combining different types of data visualization into the scene structure offered by the library.

It is worth asserting that this collection of images does not intend to be exhaustive. In Warburgian spirit, the images collected are part of a lived collecting experience, sometimes intentional but sometimes fortuitous, of one of the authors of this paper; this is specially the case in the post nineteenth century images like memes, parodies, and other types of reappropriations of the Cargeros image. Following this line, we do not see the tools used here as large scale, number crunching machines that extract general patterns of a representative dataset, but as interfaces that facilitate the process of subjective collection, interpretation, and presentation by the researchers, on one hand, and as tools for exploring arguments visually and interactively by users, allowing them to produce their own interpretations. Following Drucker, we want to make explicit the act of subjective reading, and to use the tool not just for representing, but also for modelling interpretation: “[t]he idea of modelling implies that a graphical expression serves as a primary mode of knowledge production, not a secondary expression of preexisting data. This suggests that a graphical expression might be used to create and/or show the features of a model of interpretative activity” (Drucker 2020, 2). In this sense, our method intends to integrate visual representation, interactivity, and textuality into a historic argument.

2) Next in the process, we used the filtering function of dictionary-like data structures, available in programming languages like Python and Javascript, to create subsets of our main dataset based on conditionals applied over different columns of temporal and categorical data; the concrete filters used for the Cargueros dataset are described in detail in the next section. We understand conditional filtering as a way to interpret, in the sense of partially focusing on particular elements of interest in a group of objects to get insights that would not be apparent in a space with noisier, competing information. Thus, we use filtering both as a search strategy but also as an argumentative artifact; searching and focusing is useful to make connections evident and to make patterns more salient. This approach resonates with the subbranch of media studies called “software studies”, which intends to understand and make explicit how computational operations (for instance, loops, randomization, conditionals, functions, etc.) are culturally situated and become epistemological instruments and strategies.

3) We then used these filtered data subsets to create multiple graphics based on three types of data visualizations, these graphics are our own attempt at computationally generating Warburgian panels: circle packed tree maps, used to represent nested categorical hierarchies; 2D scatterplots, used to represent relationships between two series of data in a cartesian plane; and simple adjacent dispositions of images, used to show particular details in pairs of images that deserve attention. With these three types of visualizations, we intend to show different aspects of our dataset: hierarchies and groupings, changes through time, numeric correlations, visual commonalities or differences. 4) Finally, for creating an interactive story in Aventura, one must describe a set of scenes in a JavaScript Object. For each scene, it is possible to define interface elements, like the texts, images, buttons, and clickable areas, and hypertext elements —namely, to which other scenes it is possible to jump from the current scenes. With this information, Aventura creates an interface that lets the user move between scenes. We propose that this hypertextual system is useful to create a Z-dimension in which helps to move from one panel to another to further deepen the reading of the dataset and its iconographic relations. This hypertextual Z-dimension is in essence non-lineal, as the user can make decisions about which other panels or narrative paths to explore. Non/linearity allows the user to get into separate arguments or readings of the same dataset or to move from distant reading into close reading of the objects represented. We defined a set of scenes based on particular readings, available data and historic arguments about the image of the Carguero: . These readings are described in more detail in the following section. For combining the default scene system of Aventura with the requirements of this project, we wrote two functions that integrate and preprocess the conditional filtering strategy and the data visualizations before the scenes are rendered. These ad-hoc functions are currently a prototype that could eventually be fused into the main code of Aventura to expand its functionality.


```python
# Import dependencies

import pandas as pd
from IPython.display import HTML
from observable_jupyter import embed
```

```python
# Data set

data_path = "https://docs.google.com/spreadsheets/d/1xkt2R62trJRBdELOjTXikNjl-sC0aR4R6ojQHygH6wc/export?format=csv"
df = pd.read_csv(data_path)
df['url'] = "https://lh3.googleusercontent.com/d/" + df['id']
csv_df = df.to_csv(index=False)
df.head()
```

```python
# The scenes that describe the structure of the interactive panels

scenes = {
    "1": {
      "viz": {"filter": [["periodo", "<", "1819 - 1850"]], "type": "pack", "x": "modelo", "y": "figura/fondo"},
      "options": [
        { "btn": "pack", "scene": "2" },
        { "btn": "compare", "scene": "3" },
        { "btn": "scatter", "scene": "4" }
      ]
    },
    "2": {
      "viz": {"filter": [["fecha", ">", 2000]], "type": "pack", "x": "figura/fondo", "y": "modelo"},
      "options": [
        { "btn": "cont..", "scene": "1" },
      ]
    },
    "3": {
      "viz": {"filter": [], "type": "compare", "x": 0, "y": 1},
      "options": [
        { "btn": "cont..", "scene": "1" },
      ]
    },
    "4": {
      "viz": {"filter": [["fecha", "<", 1900]], "type": "scatter", "x": "figura/fondo", "y": "modelo"},
      "options": [
        { "btn": "pack", "scene": "2" },
        { "btn": "compare", "scene": "3" }
      ]
    }
  }
```

```python
# Process the scenes and render the interactive panel
embed('@srsergiorodriguez/paneles_ed',
      cells=['panelContainer', 'processedScenes'],
      inputs={'csvString': csv_df, 'scenes': scenes}
     )
```

## Results

We can divide our collection into two large groups of images: pre 19th century images, which have been collected exhaustively , and reinterpretations created post 20th century (namely, remixes, artworks, memes, and other intertextual works); for this second group, the images collected are representative of possible use cases, but are by no means exhaustive of the multiplicity of works circulating in popular culture. For the narrative account we propose on this paper, a filter was created to obtain only images until the 19th century (“date < 1900”). In particular, we were interested in visualizing the tagged categories “figure/background” (in Spanish, “figura/fondo”), “source model” (“modelo”), “date” (“fecha”) and “period” (“periodo”). The "Figure/background" category indicates the proportion between figure and background in the image in this way: the images that predominantly represent landscapes without human figures are labeled as "background" (“fondo”), the images that predominantly represent landscapes that contain humans are labeled as "balance" (“equilibrio”), and the images in which there is a predominance of human figures are labeled as "figure" (“figura”). On the other hand, the “source model” category groups images according to the model they are based on, we identified a set of iconographic models that were extensively used and reappropriated in subsequent images (NOMBRAR AQUI LA LISTA DE MODELOS y una descripción breve). Finally, the category "Period" was organized around periodization criteria related to the Colombian history of the 19th century: in this case we used the period "1750-1810" (transition between the Ancient Régime and the Republic); and “1819-1850” (early Republic period).
When the attribute "date" is placed on the X axis and the attribute "figure/background" is placed on the Y axis, it can be seen that there are more images labeled as "background" or as "balance" in the period between 1770 and 1810. And for its part, after the period 1819-1850 onwards one can notice a preponderance of the "figure" over the "background" and the "balance.” Additionally, it is worth noting that the most used model during the period “1819 – 1850” was François Desiré Roulin. This can be seen in the scatter plot by contrasting “date < 1900” on the “y-axis” with the “model” on the “x-axis.” It can also be seen in the Circle Pack Tree Map when filtering by the period "1819-1850,” and displaying the variables "source model" and "background and figure.”
It is striking that this is the most used model, since all travelers quote Humboldt as a reference for the description of Carguero, but they do not use his image as a model. It can be seen in a contrast graph of two adjacent images that there are differences in style. It is especially important to note that Humboldt's image highlights the Carguero as part of the landscape, whereas Roulin’s highlights its physical characteristics. When viewing the TreeMap with the “source model” attribute filtered by "Humboldt,” and contrasting the "period" attribute with the "figure/background" attribute, it can be seen that, in the "1819-1850" period, the images which used Humboldt’s image as a model turned the “landscape” into “figures,” and others took the “figures” out and turned the images into “background.”
Then, the following question arises: why did the travelers who were in “Colombia/Nueva Granada” in the period “1819-1850” not use Humboldt’s model? By reviewing the primary sources and secondary literature, we conclude that two factors must be taken into account to answer the question.
First, Roulin's conceptual and visual universe is directly related to his scientific training as a physiologist practicing comparative anatomy in the scientific circle of the Academy of Natural Sciences in Paris. Roulin traveled to Colombia as a member of a scientific expedition that the plenipotentiary minister Francisco Antonio Zea had managed before personalities such as Geroges Cuvier and Alexander von Humboldt. Likewise, he published some of the results of his research in the Annals of Natural Sciences and participated in the posthumous edition of the Animal Kingdom created by Cuvier's disciples in 1836. Roulin's watercolors and drawings show his interest in highlighting aspects related to the animal and human classification that had been postulated in this scientific circle, which, from a visual point of view, sought to establish relationships between external characters, with the systematic functioning of internal organs. In the specific classification of the human race as a bipedal mammal, Cuvier considered that its development was determined by external aspects, such as climate or geography, and internal aspects, such as the intellectual characteristics of the races. These were organized in a hierarchy that descended from the Caucasian race to the Ethiopian, which was the least developed. The position of Americans in this classification was uncertain, since it was not possible to resolve the multiplicity of characteristics observed in the same category. Travelers like Roulin, and later Alcide d'Orbigny, made observations to try to solve this problem. As an analytical tool, they transposed the category of "type" to find a way to organize the multiplicity of human populations existing in America. In this sense, Roulin's Carguero drawing can be interpreted as an investigation into a “social type” that existed because of specific climatic, geographical, and racial features. These are rationalized by the observations of the external characters and their relations with internal organic functioning. Roulin refers to the Carguero as a "human mount", allowing comparison of different modes of transportation and uses of human and animal muscle-driven motion power.
Second, the drawings and images made by Roulin participated in a network of image circulation related to 19th century visual culture, such as the rationalization of production, division of labor, and mass production for a market of travelers who were interested in these products. Researchers have compared collections of images belonging to various travelers who were in the period, including the corpus of images attributed to Roulin, and have found a series of iconographic coincidences that indicate the existence of practices of adapting models to different contexts by means of details, such as the background, tools, clothing, costumes, and other elements that vary between the compositions. These practices are analyzed by researchers through notions such as "multiple authorship" and "multiple originals", which allows them to be placed on the horizon of other image-making practices that took place throughout South America in the postcolonial hispanic space
In this way, travelers like John Potter Hamilton (1827), Alcide D'Orbigny (1841), Cesar Famin (1837) and M. A. de Latrre (1848), adapted Roulin's Carguero drawing for their travelogues. Each case was adapted to its own visual and conceptual universe. Although all quote Humboldt when recounting their encounter with the Carguero, they preferred Roulin's drawing as a model that fit a growing interest in exploring American human types.
